```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
# load fusions.csv 
fusions = pd.read_csv('fusions.csv', index_col=0)
fusions.head()
```

```python
fusions_imp = fusions[['t_voix_exprimes_t2', 'sum_scores', 'inscrits','nb_in_fusion','prop_index']]
fusions_imp.head()
```

```python
# plot (sum_scores-t_voix_exprimes_t2) vs prop_index 
# size of the dots according to "inscrits"

plt.scatter( fusions_imp['prop_index'], fusions_imp['t_voix_exprimes_t2']-fusions_imp['sum_scores'], s=fusions_imp['inscrits']/1000, alpha=0.5)
plt.ylabel('sum_scores - t_voix_exprimes_t2')
plt.xlabel('prop_index')
plt.title('sum_scores - t_voix_exprimes_t2 vs prop_index')
# horizontal line at y=0
plt.axhline(0, color='red', linestyle='--')
plt.xlim(0)
plt.show()
```

```python
plt.scatter( fusions_imp['prop_index'], fusions_imp['t_voix_exprimes_t2']/fusions_imp['sum_scores'], s=fusions_imp['inscrits']/1000, alpha=0.5)
plt.ylabel('t_voix_exprimes_t2 / sum_scores')
plt.xlabel('prop_index')
plt.title('t_voix_exprimes_t2 / sum_scores vs prop_index')
# horizontal line at y=1
plt.axhline(1, color='red', linestyle='--')
plt.xlim(0)
plt.show()
```

```python
# same but only if nb_in_fusion = 2 
fusions_imp_2 = fusions_imp[fusions_imp['nb_in_fusion'] == 2]
plt.scatter( fusions_imp_2['prop_index'], fusions_imp_2['t_voix_exprimes_t2']-fusions_imp_2['sum_scores'], s=fusions_imp_2['inscrits']/1000, alpha=0.5)
plt.axhline(0, color='red', linestyle='--')
plt.xlim(0)
plt.ylabel('sum_scores - t_voix_exprimes_t2')
plt.xlabel('prop_index')
plt.title('sum_scores - t_voix_exprimes_t2 vs prop_index (nb_in_fusion = 2)')
plt.show()
```

```python
# same but only if nb_in_fusion = 2 
fusions_imp_2 = fusions_imp[fusions_imp['nb_in_fusion'] == 2]
plt.scatter( fusions_imp_2['prop_index'], fusions_imp_2['t_voix_exprimes_t2']/fusions_imp_2['sum_scores'], s=fusions_imp_2['inscrits']/1000, alpha=0.5)
plt.axhline(1, color='red', linestyle='--')
plt.xlim(0)
plt.ylabel('t_voix_exprimes_t2 / sum_scores')
plt.xlabel('prop_index')
plt.title('t_voix_exprimes_t2 / sum_scores vs prop_index (nb_in_fusion = 2)')
plt.show()
```
